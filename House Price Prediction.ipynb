```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
```

```python
df = sklearn.datasets.load_boston()
```

```python
df
```

```python
# loading the dataset to a pandas df
df1 = pd.DataFrame(df.data, columns = df.feature_names)
```

```python
df1 
```

```python
#Add the target column to the dataframe
df1['price'] = df.target
```

```python
df1
```

```python
df1.info()
```

```python
df1.describe()
```

```python
corr = df1.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8},cmap='Blues')
```

```python
#cheking missing values
df1.isnull().sum()
```

```python
#splitting the data and target features
X = df1.drop(columns="price", axis=1)
y= df1["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

```python
#XGBoost Regressor
model =XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

Evalaution


```python
#Training data accuracy
prediction =model.predict(X_train)
```

```python
#R squared error
score_1 = metrics.r2_score(y_train,prediction)
#MAE
score_2 = metrics.mean_absolute_error(y_train,prediction)
```

```python
print('R squared error:', score_1)
print('Mean absoulte erroe:', score_2)
```

visualize the actual and predicted prices

```python
plt.scatter(y_train, prediction)
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
#Testing data evolution
prediction =model.predict(X_test)
```

```python
#R squared error
score_1 = metrics.r2_score(y_test,prediction)
#MAE
score_2 = metrics.mean_absolute_error(y_test,prediction)
#RMSE
score_3 = metrics.mean_squared_error(y_test,prediction)
```

```python
print('R squared error:', score_1)
print('Mean absoulte erroe:', score_2)
```
